## Handling Categorical Data 
### models prefer to deal with numbers rather than words if possible, that's what we are going to do now 
## 1- "Manufacturer"

```python
# import the needed librarys
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
features_matrix['Manufacturer'] = encoder.fit_transform(features_matrix['Manufacturer'])
features_matrix
```

## Fuel Type

```python
# import the needed librarys
features_matrix['Fuel Type'] = encoder.fit_transform(features_matrix['Fuel Type'])
features_matrix
```

```python
goal_vector = encoder.fit_transform(goal_vector)
goal_vector
```

## a good way to handle such case, is to use the one hot encoder

### Now let's have our data encoded in a dummy variables, so we no longer have the problem of one country or(category) taking a higher value than the other

```python
# import the oneHotEncoder class
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder(categorical_features=[3])
features_matrix = oneHotEncoder.fit_transform(features_matrix).toarray()
features_matrix[:, :3]
```

## Now we need to split our data to train and test data ..

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_matrix, goal_vector, test_size=.3, random_state=0)
model = LinearRegression()
model.fit(x_train, y_train)
```

## Perdiction 

```python
print(mean_squared_error(y_train, model.predict(x_train)))
model.score(x_train, y_train)
```

```python
print(mean_squared_error(y_test, model.predict(x_test)))
model.score(x_test, y_test)
```
